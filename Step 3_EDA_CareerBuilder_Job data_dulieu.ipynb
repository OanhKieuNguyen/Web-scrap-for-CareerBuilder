```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from unidecode import unidecode
```

```python
df = pd.read_csv('D:\WORK\Data Analysis\Raw data\Data-Analyst-Clean.csv', encoding='utf-16')
display(df.head())
print(df.columns)
```

```python
old_col_name=df.columns.copy()
new_col_names=[]
for col in df.columns:
    new_col_names.append(unidecode(col).lower().replace(' ','_').replace('/','_').replace('-','_'))
df.columns=new_col_names
df.columns
```

```python
df.shape
```

```python
location_cols = df.columns[df.columns.str.contains("location_")]
industry_cols = df.columns[df.columns.str.contains("industry_")]
```

```python
today = '03/12/2023'
```

```python
sns.set_style('darkgrid')
loc_count = {}
for location_col in location_cols:
    loc_count.update({location_col: df.query(f'`{location_col}` == 1').shape[0]})

loc_count_df = pd.DataFrame.from_dict(loc_count, orient='index', columns=['count'])
loc_count_df.nlargest(10, columns='count').sort_values(by='count').plot(kind='barh')

plt.title(f'number of jobs per location (top 10) - update {today}')
plt.xlabel('count')
plt.tight_layout()
plt.show()
```

```python
industry_count = {}
for industry_col in industry_cols:
    industry_count[industry_col] = df[industry_col].sum() # 2nd way to count, this is cleaner than 👇. summing bool variable is a nice trick for counting. 
    # industry_count.update({industry_col: df.query(f'{industry_col} == 1').shape[0]})

industry_count_df = pd.DataFrame.from_dict(industry_count, orient='index', columns=['count'])
# draw plot for top 10
industry_count_df.nlargest(10, 'count').sort_values(by='count').plot(kind='barh')

plt.title(f'number of jobs per industry (top 10) - update {today}')
plt.xlabel('count')
plt.tight_layout()
plt.show()
```

```python
df.query('job_level != "failed"').job_level.value_counts().sort_values().plot(kind='barh')
plt.title(f'number of job per job_level - update {today}')
plt.tight_layout()
```

```python
# We will analyze deeper for banking industry
df_bank = df[df['industry_ngan_hang']==1]
df_bank
```

```python
location_cols
```

```python
sns.set_style('darkgrid')
loc_count = {}
for location_col in location_cols:
    loc_count.update({location_col: df_bank.query(f'{location_col} == 1').shape[0]})

loc_count_df = pd.DataFrame.from_dict(loc_count, orient='index', columns=['count'])
top_10_locations = loc_count_df.nlargest(10, columns='count').sort_values(by='count')
top_10_locations.index=top_10_locations.index.str.replace('location_', '')

ax = top_10_locations.plot(kind='barh')
plt.title(f'Top 10 thành phố nhiều jobs banking nhất - update {today}')
plt.xlabel('count')
plt.tight_layout()

# Add value labels on top of the bars
for i, v in enumerate(top_10_locations['count']):
    ax.text(v+0.5, i, str(v), color='black', va='center')

plt.show()
```

Most banking jobs are in Ha Noi as of 3/12/2023 data. It is quite matched with situation that Head Office of Vietnamese bank is mostly located in Ha Noi

```python
job_titles=pd.Series(df_bank.job_title.unique())
for i in job_titles: 
    print(i)
```

```python
job_titles.shape
```

```python
#job_titles=pd.Series(df_bank.job_title.unique())
# for i in job_titles: 
#     print(i)
keywords=['phân tích', 'analyst', 'analysis']
DA_jobs=[]
for i in keywords:
    for job in job_titles[job_titles.str.contains(i)]:
        if job not in DA_jobs:
            DA_jobs.append(job)
print(len(DA_jobs))
job_titles[~job_titles.isin(DA_jobs)]
```

```python
DA_jobs
```

```python
keywords=['kỹ sư', 'engineer']
DE_jobs=[]
for i in keywords:
    for job in job_titles[job_titles.str.contains(i)]:
        if job not in DE_jobs:
            DE_jobs.append(job)
print(len(DE_jobs))
job_titles[~job_titles.isin(DE_jobs)][~job_titles.isin(DA_jobs)]
```

```python
keywords=['khoa học', 'scientist','mô hình','ai','ml']
DS_jobs=[]
for i in keywords:
    for job in job_titles[job_titles.str.contains(i)]:
        if job not in DS_jobs:
            DS_jobs.append(job)
print(len(DS_jobs))
job_titles[~job_titles.isin(DE_jobs)][~job_titles.isin(DA_jobs)][~job_titles.isin(DS_jobs)]
```

```python
# df_bank[df_bank.job_title]
df_bank['job_group']='others' # this is just like doing fillna() before hand
df_bank.loc[df_bank[df_bank.job_title.isin(DA_jobs)].index, 'job_group']='DA'
df_bank.loc[df_bank[df_bank.job_title.isin(DE_jobs)].index, 'job_group']='DE'
df_bank.loc[df_bank[df_bank.job_title.isin(DS_jobs)].index, 'job_group']='DS'
df_bank[['job_title', 'job_group']]
```

```python
# def exp_min_cal(city):
aggfunc_dict = {
    'sal_min': 'mean',
    'sal_max': 'mean',
    'job_id': 'count'
}
print(
    f"\033[1;31mTable 1: Lương theo số năm kinh nghiệm Min\033[0m")
pivot_table = pd.pivot_table(data=df_bank, index='exp_min', values=[
                                'sal_min', 'sal_max', 'job_id'], aggfunc=aggfunc_dict).reset_index()[['exp_min', 'sal_min', 'sal_max', 'job_id']].rename(columns={'job_id':'count'})
display(pivot_table
        .style.format("{:,.1f}", subset=['sal_min', 'sal_max', 'exp_min'])
        .background_gradient(cmap='Blues', vmin=0, vmax=50, subset=['sal_min', 'sal_max'])
        )

plt.bar(x=pivot_table.exp_min, height=pivot_table.sal_min)
plt.title(f"Lương theo số năm kinh nghiệm MIN - update {today}")
plt.xticks(ticks=pivot_table.exp_min)
plt.xlabel('year of experience')
plt.ylabel('salary mininum (million VND)')
# Add labels on top of each bar
for x, y in zip(pivot_table.exp_min, pivot_table.sal_min):
    plt.text(x=x, y=y+0.3, s=f"{y:,.1f} Tr", ha='center')
# plt.tight_layout()
plt.show()


# # top_10_loc=pd.Series(loc_count_df.nlargest(10, columns='count').index).str.replace('location_', '')
# lkp_cities = [
#     'ho_chi_minh',
#     'ha_noi',
#     'hai_phong',
#     'binh_duong',
#     'da_nang',
#     # 'dong_nai',
#     # 'bac_ninh',
# ]
# for i in lkp_cities:
#     exp_min_cal(i)
```
